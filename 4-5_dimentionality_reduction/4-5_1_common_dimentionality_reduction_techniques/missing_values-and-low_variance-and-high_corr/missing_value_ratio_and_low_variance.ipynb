# Missing Value Ratio
## Formula:
### Ratio of missing value = no. of missing values / total no. of observations

# Low Variance Code Technique 2:

```python
data = pd.read_csv("E:\Machine_Learning_Journey\Machine_Learning_Journey\data_sets\Raw_Housing_Prices.csv")
data.head()
```

```python
data.info()
```

```python
numerical_data = data.drop(columns = ['ID', 'Date House was Sold', 'Sale Price',
                               'Waterfront View', 'No of Times Visited',
                            'Condition of the House', 'Overall Grade', 'Zipcode'
                              ])

numerical_data.info()
```

```python
from sklearn.preprocessing import Normalizer
from sklearn.impute import SimpleImputer
# Step 1: Handle missing values using SimpleImputer
imputer = SimpleImputer(strategy='mean')  # You can use 'median', 'most_frequent', or 'constant'
numerical_data_imputed = pd.DataFrame(imputer.fit_transform(numerical_data), columns=numerical_data.columns)

# Step 2: Apply normalization
norm = Normalizer()
norm_data = norm.fit_transform(numerical_data_imputed)

# Step 3: Convert the normalized data back to a DataFrame
norm_data = pd.DataFrame(norm_data, columns=numerical_data.columns)

# Display the first few rows of the normalized data
norm_data.head()
```

```python
tmp = norm_data.var()
tmp
```

```python
T = 1e-04

for i in tmp:
  if i < T:
    norm_data.drop(columns = [tmp[tmp ==i].index[0]], inplace = True)

norm_data.var()
```
