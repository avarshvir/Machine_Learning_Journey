```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Age' : [25,30,35,40,45],
    'Income' : [50000,60000,75000,90000,80000],
    'Education' : ['High School', 'Bachelor', 'Master', 'ph.D.', 'Bachelor'],
    'City' : ['New York','San Francisco', 'Chicago', 'Los Angeles', 'Miami'],
    'Gender' : ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Productivity' : [5,4,3,2,4]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df_encoded = pd.get_dummies(df, columns = ['Education', 'City', 'Gender'])
```

```python
df_encoded
```

## feature scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df_encoded[['Age','Income']]
```

```python
scaled_feature = scaler.fit_transform(df_encoded[['Age','Income']])
```

```python
scaled_feature
```

```python
df_encoded[['Age','Income']] = scaled_feature
```
